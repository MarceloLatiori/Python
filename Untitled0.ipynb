```python
# variáveis globais
lista_livro = []
id_global = 0

# função para cadastrar livros
def cadastrar_livro(id):
    print('Bem-vindo ao cadastro de livros de Marcelo Latiori')
    print('Código do produto: {}'.format(id))
    nome_do_livro = input('Digite o nome do livro: ')
    autor_do_livro = input('Digite o nome do autor do livro: ')
    editora_do_livro = input('Digite o nome da editora: ')
    dicionario_livro = {'id': id, 'nome_do_livro': nome_do_livro, 'autor_do_livro': autor_do_livro, 'editora_do_livro': editora_do_livro}
    lista_livro.append(dicionario_livro)
    print('Livro cadastrado com sucesso!')

# função para consultar livros
def consultar_livro():
    print('Bem-vindo à consulta de livros de Marcelo Latiori')
    while True:
        consultar = input('Escolha a consulta desejada: \n'+
                          '1 - Consultar todos os livros: \n'+
                          '2 - Consultar por id: \n'+
                          '3 - Consultar pelo Autor: \n'+
                          '4 - Retornar ao Menu: \n')

        if consultar == '1':
            print('Consultar todos os livros')
            for livro in lista_livro:
                for chave, valor in livro.items():
                    print('{}:{}'.format(chave, valor))
        elif consultar == '2':
            print('Consultar por id')
            id_livro = int(input('Entre com o id desejado: '))
            for livro in lista_livro:
                if livro['id'] == id_livro:
                    print('-----------')
                    for chave, valor in livro.items():
                        print('{}:{}'.format(chave, valor))
        elif consultar == '3':
            print('Consultar pelo Autor ')
            autor_desejado = input('Entre com o nome do Autor: ')
            for livro in lista_livro:
                if livro['autor_do_livro'] == autor_desejado:
                    print('-----------')
                    for chave, valor in livro.items():
                        print('{}:{}'.format(chave, valor))
        elif consultar == '4':
            print('Você escolheu retornar ao menu')
            break
        else:
            print('Entre com um Serviço Válido')

# função para remover livros
def remover_livro():
    print('Bem-vindo à remoção de livros de Marcelo Latiori')
    id_livro = int(input('Qual o id do Livro que deseja remover: '))
    for livro in lista_livro:
        if livro['id'] == id_livro:
            lista_livro.remove(livro)
            print('Livro removido com sucesso!')
            return
    print('Livro não encontrado.')

# Estrutura de menu no main
while True:
    cadastro = input('Escolha o serviço desejado: \n'+
                      '1 - Cadastrar Livro: \n'+
                      '2 - Consultar Livro: \n'+
                      '3 - Remover Livro: \n'+
                      '4 - Retornar ao Menu: \n'+
                      '5 - Encerrar: \n')
    if cadastro == '1':
        id_global += 1
        cadastrar_livro(id_global)
    elif cadastro == '2':
        consultar_livro()
    elif cadastro == '3':
        remover_livro()
    elif cadastro == '4':
        print('Retornando ao menu.')
    elif cadastro == '5':
        print('Encerrando o programa.')
        break
    else:
        print('Entre com um Serviço Válido')
```
